```python
def datagen() : #number
    # 랜덤 시드 고정시키기
    np.random.seed(3)
    filename_in_dir = []
    datagen = ImageDataGenerator(rescale=1./255)
    datagen = ImageDataGenerator(rescale=1./255,
                                 rotation_range=30,
                                 width_shift_range=0.3,
                                 height_shift_range=0.3,
                                 brightness_range=(0.0, 0.9),
                                 shear_range=0.7,
                                 zoom_range=[0.9, 2.2],
                                 horizontal_flip=True,
                                 vertical_flip=True,
                                 fill_mode='nearest')
    input_dir = currentPath+"/OpenSource/train_dir/"#+number
    input_files = os.listdir(os.path.join(input_dir))
    
    for files in input_files :
        full_fname = os.path.join(input_dir, files)
        filename_in_dir.append(full_fname)
        
    for file_image in filename_in_dir :
        print(file_image)
        img = load_img(file_image)
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape)
        title, ext = os.path.splitext(file_image)
        print("title : "+title+"\t ext : "+ext)
        i = 0
        for batch in data_datagen.flow(x, save_to_dir = input_dir, save_prefix=title, save_format='jpg') :
            i += 1
            if i>1000 :
                break
```
